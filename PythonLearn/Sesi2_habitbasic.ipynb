```python
# Get a list as input from user in Python List of lists as input
lst = []
n = int(input("Enter number of elements : "))

for i in range(0, n): #iterasi jumlah perulangan
    ele = [input(), int(input())] #inisiasi dalam elemen list sesuai jumlah perulangan
    lst.append(ele)

print(lst)
# ini intinya ingin menduplikasi sebuah list ke dalam list
```

```python
def createList(r1, r2):

 # Testing if range r1 and r2
 # are equal
 if r1 == r2:
  return r1

 else:

  # Create empty list
  res = []

  # loop to append successors to
  # list until r2 is reached.
  while r1 < r2+1:

   res.append(r1)
   r1 += 1 # increment buat mengatur isi elemnt dan polanya.
  return res

def filter(wadah, r1, r2): #ngide ini gabut
  new_range = []
  for x in wadah:
    if r1 < x < r2:
      new_range.append(x)
  print(new_range)

# Driver Code
r1, r2 = -1, 3
z = createList(r1, r2)
print(z)
filter(z, 0, 3)
```

```python
# IF Else Statement
num = 5
if num >= 0:
    print("Positive or Zero")
else:
     print("Negative number")
```

```python
# IF, Elif, Else Statement
num = 0
if num > 0:
    print("Positive number")
elif num == 0:
    print("Zero")
else:
    print("Negative number")
```

```python
# IF nested
num = 8
if num >= 0:
    if num == 0:
        print("zero")
    else:
        print("Positive number")
else:
    print("Negative number")
```

# Loops

```python
# While Loop
num = 1
odd_nums = []
while num:
    if num % 2 != 0:
        odd_nums.append(num)
    if num >=20:
        break
    num += 1
print("Odd numbers: ", odd_nums)
```

```python
# For Loop
list = [1, 2, 3, 4, 5]
for num in list:
    print(num)
```

```python
print("Python for loop with zip")
names = ["Alice", "Bob", "Charlie"]
scores = [90, 85, 88]
for name, score in zip(names, scores): #zip method bertujuan untuk menyatukan 2 var list menjadi key : value yang jika dengan loop maka setiap elemen var list 1 menjai key dan var list 2 sebagai valuenya
    print(f"{name}: {score}")
```

```python

print("\n Python for loop with yield")
def countdown(n):
    while n > 0:
        yield n #yield berguna untuk mengembalikan nilai n ke fungsi untuk dipanggil kembali
        n -= 1

for num in countdown(5): #method sebagai modal/input loop for
    if num == 2:
        print("cek")
    print(num)
```
