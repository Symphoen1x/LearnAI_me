The __call__ method in Python is a special method that allows an instance of a class to be called as if it were a function. When you define the __call__ method in a class, you can use the instances of that class as callable objects.

Here's a breakdown of how it works:

Defining the __call__ Method
The __call__ method is defined within a class like any other method. The first parameter is always self, which refers to the instance of the class. Additional parameters can be defined based on what the method needs to do.

more: https://chatgpt.com/c/8ff7d4ee-11e2-4fb6-b282-351449a9d008 

```python
# __call__ method implemen
class MyCallableClass:
    def __init__(self, value):
        self.value = value

    def __call__(self, new_value):
        print(f"Old value: {self.value}")
        self.value = new_value
        print(f"New value: {self.value}")

# Create an instance of the class
my_instance = MyCallableClass(10)

# Call the instance as if it were a function
my_instance(20)
```

### Apa perbedaan penggunaan @staticmethod, constructor method, dan @clashmethod  ?
ada 3 perbedaan utama sebagai aspek yang harus dipertimbangkan:<br>
**segi parameter**:  <br>
* @staticmethod: Tidak menerima cls atau self.
* constructor method: menerima argumen bebas sbg ketetapan, tapi lebih sering self.
* @classmethod: Menerima cls.

**segi akses**: <br>
* @staticmethod: Tidak dapat mengakses atau memodifikasi atribut kelas atau instance.
* constructor method: Dapat mengakses dan memodifikasi atribut kelas.
* @classmethod: Dapat mengakses dan memodifikasi atribut kelas.

**segi penggunaan**: <br>
* @staticmethod: Digunakan untuk method yang tidak tergantung pada status kelas atau instance. biasanya method khusus untuk tujuan tertentu.
* @classmethod: Digunakan untuk method yang perlu mengetahui atau memodifikasi variabel kelas.
* constructor method: Digunakan untuk method yang perlu mengetahui atau memodifikasi variabel instance.


```python
# contoh implemen @staticmethod
class MathOperations:
# Tidak dapat mengakses atau memodifikasi atribut kelas atau instance.
    @staticmethod
    # Tidak menerima cls atau self.
    def add(x, y):
        return x + y

# Memanggil metode statis tanpa membuat instance dari kelas
result = MathOperations.add(5, 3)
print(result)  # Output: 8
```

```python
# implemen @classmethod
class MathOperations:
    factor = 2
    
    @classmethod
    # Dapat mengakses dan memodifikasi atribut kelas
    def multiply(cls, x):
        return x * cls.factor
    
    @classmethod
    # Digunakan untuk metode yang perlu mengetahui atau memodifikasi status kelas. Bisa juga untuk dimodifikasi dimethod lain ketika memanggil method lain.
    def set_factor(cls, new_factor):
        cls.factor = new_factor

# Memanggil metode kelas dari kelas
result = MathOperations.multiply(5)
print(result)  # Output: 10

# Mengubah nilai faktor
MathOperations.set_factor(3)
result = MathOperations.multiply(5)
print(result)  # Output: 15
```

### Apa itu pass keyword dan fungsinya dalam class atau method class?
* tentunya dalam class atau method itu, si pass ini bisa untuk template yang ada di anak class.
* bisa juga untuk menjalankan code lain yang diprioritaskan, tapi takut method ini hilang dari susunan list inisasi method awal sehingga walaupun dia ada dan nyala, tapi tidak bervalue/mengganggu/beroutput.

```python
# contoh sbg template
class Shape:
    def __init__(self, name):
        self.name = name

    def area(self):
        pass  # Metode ini akan diimplementasikan di subclass

    def perimeter(self):
        pass  # Metode ini akan diimplementasikan di subclass

# Subclass yang mengimplementasikan metode area dan perimeter
class Circle(Shape):
    def __init__(self, radius):
        super().__init__("Circle")
        self.radius = radius

    def area(self):
        import math
        return math.pi * self.radius ** 2

    def perimeter(self):
        import math
        return 2 * math.pi * self.radius

# Contoh penggunaan
c = Circle(5)
print(f"Area of {c.name}: {c.area()}")
print(f"Perimeter of {c.name}: {c.perimeter()}")
```

### Apa perbedaan method str() dan repr()
sumber: https://realpython.com/python-repr-vs-str/ 
__str__
* Tujuan: Memberikan representasi string yang mudah dibaca oleh manusia.
* Penggunaan: Digunakan ketika Anda menginginkan representasi objek yang ramah pengguna, biasanya untuk output pengguna akhir.
* pemanggilan: bisa dengan print variabel 
__repr__
* Tujuan: Memberikan representasi string yang lebih rinci dan formal yang dimaksudkan untuk debugging dan pengembangan. Representasi ini harus, jika memungkinkan, menjadi ekspresi valid yang dapat digunakan untuk merekonstruksi objek yang sama.
* The reason there are two methods to display an object is that they have different purposes:

.__repr__() provides the official string representation of an object, aimed at the programmer. 

.__str__() provides the informal string representation of an object, aimed at the user. 
* Penggunaan: 
.__repr__() : Digunakan ketika Anda membutuhkan representasi yang lebih teknis dan rinci dari objek, seperti dalam konteks debugging atau logging.
.__str__() : Digunakan ketika Anda membutuhkan representasi yang lebih sederhana dan formal dari objek, seperti dalam output pengguna akhir. Ini akan lebih mudah dipahami pengguna. 
* pemanggilan: bisa dengan print variabel 


```python
# versi code from python
import datetime
today = datetime.datetime.now()
#for programmer
repr(today)
'datetime.datetime(2023, 2, 18, 18, 40, 2, 160890)'
today.__repr__()
'datetime.datetime(2023, 2, 18, 18, 40, 2, 160890)'
```

```python
# for user
str(today)
'2023-02-18 18:40:02.160890'
today.__str__()
'2023-02-18 18:40:02.160890'
```

```python
# versi manual
class Book:
    def __init__(self, title, author):
        self.title = title
        self.author = author

    def __repr__(self):
        class_name = type(self).__name__
        return f"{class_name}(title={self.title!r}, author={self.author!r})"

    def __str__(self):
        return f'"{self.title}" by {self.author}'

odyssey = Book("The Odyssey", "Homer")

print(repr(odyssey))
print(str(odyssey))
```

```python
class Animal:
    def sound(self):
        print("Some generic animal sound")

class Dog(Animal):
    def sound(self):
        # Memanggil method sound dari kelas induk (Animal)
        super().sound()
        print("Bark")

# Membuat objek Dog dan memanggil method sound
dog = Dog()
dog.sound()
```

```python
class OuterClass:
    class InnerClass:
        def display(self):
            print("Inner class method")

    def __init__(self):
        self.inner = self.InnerClass()

    def show_inner(self):
        self.inner.display()

# Membuat objek OuterClass dan memanggil method dari InnerClass
outer = OuterClass()
outer.show_inner()
```

```python
class Engine:
    def start(self):
        print("Engine started")

class Car:
    def __init__(self, engine):
        self.engine = engine  # Komposisi: Car memiliki objek Engine

    def start_engine(self):
        self.engine.start()  # Memanggil metode dari objek Engine

# Contoh penggunaan
engine = Engine()
car = Car(engine)
car.start_engine()
```

```python
class NLPProcessor:
    class Tokenizer:
        def tokenize(self, text):
            # Tokenization logic here
            return text.split()
        
    class Stemmer:
        def stem(self, word):
            # Stemming logic here
            return word.lower()  # A placeholder for actual stemming logic
        
    class Lemmatizer:
        def lemmatize(self, word):
            # Lemmatization logic here
            lemmas = {
                "running": "run",
                "ran": "run",
                "cars": "car",
                "children": "child"
            }
            return lemmas.get(word, word)
        
    class POSTagger:
        def pos_tag(self, tokens):
            # POS tagging logic here
            tags = {
                "This": "DT",
                "is": "VBZ",
                "an": "DT",
                "example": "NN",
                "sentence": "NN"
            }
            return [(token, tags.get(token, "NN")) for token in tokens]

    def __init__(self, use_stemming=True, use_lemmatization=True, use_pos_tagging=True):
        self.tokenizer = self.Tokenizer()
        self.stemmer = self.Stemmer() if use_stemming else None
        self.lemmatizer = self.Lemmatizer() if use_lemmatization else None
        self.pos_tagger = self.POSTagger() if use_pos_tagging else None

    def process(self, text):
        tokens = self.tokenizer.tokenize(text)
        
        if self.stemmer:
            tokens = [self.stemmer.stem(token) for token in tokens]
        
        if self.lemmatizer:
            tokens = [self.lemmatizer.lemmatize(token) for token in tokens]
        
        if self.pos_tagger:
            tokens = self.pos_tagger.pos_tag(tokens)
        
        return tokens

# Using the NLPProcessor
nlp = NLPProcessor()
processed_text = nlp.process("This is an example sentence.")
print(processed_text)

# Testing different configurations
nlp_no_stemming = NLPProcessor(use_stemming=True)
print(nlp_no_stemming.process("This is an example sentence."))

nlp_no_lemmatization = NLPProcessor(use_lemmatization=True)
print(nlp_no_lemmatization.process("This is an example sentence."))

nlp_no_pos_tagging = NLPProcessor(use_pos_tagging=True)
print(nlp_no_pos_tagging.process("This is an example sentence."))
```

```python
# dengan menerapan contoh code linkedlist single
class Node:
    def __init__(self, value):
        self.value = value
        self.next = None  # Referensi ke objek Node lain (self-referential)
        
    def add(self, value):
        if self.next is None:
            self.next = Node(value)  # Membuat objek Node baru
        else:
            self.next.add(value)  # Rekursi dengan memanggil method pada objek Node berikutnya

    def __str__(self):
        if self.next:
            return f"{self.value} -> {self.next}"
        else:
            return f"{self.value}"

# Membuat objek Node pertama
head = Node(1)
# Menambahkan node ke dalam linked list secara rekursif
head.add(2)
head.add(3)
head.add(4)

print(head)
```

### penggunaan abstrak method pada polimorfisme
* import dulu package abc, modul ABC, key abstractmethod, dan @abstractmethod.
* jadikan abc sbg input layaknya inheritance terhadap ABC class lalu define class dan method yang akan di polimorfisme.
* ABC (Abstract Base Classes) module adalah bagian dari library standar Python yang menyediakan kerangka dasar untuk mendefinisikan kelas abstrak. Kelas abstrak adalah kelas yang tidak dimaksudkan untuk diinstansiasi secara langsung, tetapi dirancang untuk menjadi basis bagi kelas lain. Kelas ini bisa berisi metode abstrak, yang harus diimplementasikan oleh subclass.
* *Dekorator abstractmethod digunakan untuk menandai metode sebagai abstrak. Metode abstrak adalah metode yang dideklarasikan di kelas abstrak tetapi tidak memiliki implementasi. Subclass dari kelas abstrak ini harus mengimplementasikan metode abstrak ini, jika tidak, Python akan memunculkan kesalahan TypeError.

```python
from abc import ABC, abstractmethod

class Animal(ABC):
    @abstractmethod
    def sound(self):
        # pass
        return NotImplemented

class Dog(Animal):
    def sound(self):
        return "Bark"

class Cat(Animal):
    def sound(self):
        return "Meow"

animals = [Dog(), Cat()]
for animal in animals:
    print(animal.sound())
```

```python
# kasus jika inheritance yang dipakai begitu kompleks, saran menggunakan super() + *kwargs + OPSIONAL=kombinasi pemanggilan konstruktor kelas induk
class base():
    def __init__(self, **kwargs):
        pass

class Base1(base):
    def __init__(self, str1, **kwargs):
        super().__init__(**kwargs)
        self.str1 = str1

class Base2(base):
    def __init__(self, str2, **kwargs):
        super().__init__(**kwargs)
        self.str2 = str2

class Base3(base):
    def __init__(self, str3, **kwargs):
        super().__init__(**kwargs)
        self.str3 = str3

class Derived(Base1, Base2, Base3):
    def __init__(self, str1, str2, str3, str4, **kwargs):
        super().__init__(str1=str1, str2=str2, str3=str3, **kwargs)
        self.str4 = str4

    def printStrs(self):
        print(self.str1, self.str2, self.str3, self.str4)

ob = Derived("A", "B", "C", "D")
ob.printStrs()


```

* Dekorator Property:
1. @property: Diletakkan di atas metode yang berfungsi sebagai getter. Ini mengubah metode tersebut menjadi sebuah property yang dapat diakses tanpa tanda kurung.

### Menggunakan 3 jenis dekorator property

```python
class Person:
    def __init__(self, name):
        self._name = name

    @property
    def name(self):
        return self._name

    @name.setter
    def name(self, value):
        if not isinstance(value, str):
            raise ValueError("Name must be a string") #salah satu cara jika lu pingin mengakali suatu sistem
        self._name = value

    @name.deleter
    def name(self):
        print("Deleting name...")
        del self._name

# Contoh penggunaan
p = Person("Alice")
print(p.name)  # Output: Alice

p.name = "Bob"
print(p.name)  # Output: Bob

del p.name  # Akan memanggil deleter dan menghapus atribut _name
# Output: Deleting name...

# print(p.name)
# Jika mencoba mengakses p.name setelah penghapusan, akan muncul AttributeError: 'Person' object has no attribute '_name'

#untuk itu coba ini:
try:
    print(p.name)
except AttributeError:
    print("Name attribute has been deleted")  
```

### Contoh beberapa kegunaan @property

**Pengecekan dan Validasi Tipe Atribut**

```python
class Person:
    def __init__(self, name):
        self._name = name

    @property
    def name(self):
        return self._name

    @name.setter
    def name(self, value):
        if not isinstance(value, str):
            raise ValueError("Name must be a string")
        self._name = value

# Contoh penggunaan
p = Person("Alice")
print(p.name)  # Output: Alice

p.name = "Bob"  # Tidak ada masalah
print(p.name)  # Output: Bob

p.name = 123  # Akan memunculkan ValueError: Name must be a string
```

**Membuat Atribut Hanya-Baca atau Hanya-Tulis**

```python
# Hanya bisa baca
class Circle:
    def __init__(self, radius):
        self._radius = radius

    @property
    def radius(self):
        return self._radius
   

# Contoh penggunaan
c = Circle(5)
print(c.radius)  # Output: 5

c.radius = 10  # Akan memunculkan AttributeError: can't set attribute
```

```python
# write only
class PasswordManager:
    def __init__(self):
        self._password = None

    @property
    def password(self):
        raise AttributeError("Password is write-only")

    @password.setter
    def password(self, value):
        self._password = value

# Contoh penggunaan
pm = PasswordManager()
pm.password = "securepassword"  # Tidak ada masalah

print(pm.password)  # Akan memunculkan AttributeError: Password is write-only
```

**Mengubah Variabel Instans Menjadi Atribut yang Dihitung Secara Dinamis:**

```python
class Rectangle:
    def __init__(self, width, height):
        self.width = width
        self.height = height
        
    @property
    def area(self):
        return self.width * self.height

# Contoh penggunaan
r = Rectangle(4, 5)
print(r.area)  # Output: 20

r.width = 6
print(r.area)  # Output: 30, area berubah sesuai perubahan width
```

**Menambahkan Perilaku Baru Tanpa Mengubah Kode Klien:**

```python
class Employee:
    def __init__(self, name, salary):
        self._name = name
        self._salary = salary #anggap ini sebuah variabel untuk salary yang tersembunyi dari user

    @property
    def salary(self): #dengan adanya method + @property, salary bisa diakses user value attributnya
        return self._salary

    @salary.setter #ini juga membuat salary bisa diubah/diedit oleh user, tapi bisa tetap di atur sistemnya biar tidak bias
    def salary(self, value):
        if value < 0:
            raise ValueError("Salary cannot be negative")
        self._salary = value

# Contoh penggunaan
e = Employee("John", 5000)
print(e.salary)  # Output: 5000

e.salary = 6000  # Tidak ada masalah
print(e.salary)  # Output: 6000

e.salary = -100  # Akan memunculkan ValueError: Salary cannot be negative
```

```python
# Python program showing
# abstract class cannot
# be an instantiation
from abc import ABC,abstractmethod

class Animal(ABC):
    @abstractmethod
    def move(self):
        pass
class Human(Animal):
    def move(self):
        print("I can walk and run")

class Snake(Animal):
    def move(self):
        print("I can crawl")

class Dog(Animal):
    def move(self):
        print("I can bark")

class Lion(Animal):
    def move(self):
        print("I can roar")

c=Animal()
```

## cara mengakali jika ingin melihat class abstract
ya dengan super()

```python
# Python program invoking a 
# method using super()
from abc import ABC

class R(ABC):
    def rk(self):
        print("Abstract Base Class")

class K(R):
    def rk(self):
        super().rk()
        print("subclass ")

# Driver code
r = K()
r.rk()
```

```python
import spacy
# class yang menyimpan teknik pemecah
class Doc:
    def __init__(self, text): 
        self.text = text     
        self.tokens = [word for word in text.split(',')]
    def __iter__(self):
        for token in self.tokens:
            yield token

class Language:
    def __init__(self, model_name): #model_name dimasukan/inputan objek language
        self.model_name = model_name

    def __call__(self, text): #lalu class ini juga membuat inputan kedua sbg text yang nanti menjadi parameter dalam class Doc ke objeknya
        return Doc(text) #anggap aja kayak lu numpang nama disini, tapi udh keproses langsung gitu ke atas. Disinilah __call__ bekerja dengan membuat objek yang menampung class ini bakalan bisa input text. cek dibawah

# area pemanggilan
objekDoc = Language('en_core_web_sm')

# Process text to create a Doc object
doc = objekDoc("""bought trying""")

for word in doc:
    print(word)
```

```python
#  A simple Python program to demonstrate
# working of iterators using an example type
# that iterates from 10 to given value
 
# An iterable user defined type
class Test:
 
    # Constructor
    def __init__(self, limit):
        self.limit = limit
 
    # Creates iterator object
    # Called when iteration is initialized
    def __iter__(self):
        self.x = 10
        return self #kenapa mengembalikan self?
 
    # To move to next element. In Python 3,
    # we should replace next with __next__
    def __next__(self):
 
        x = self.x #cuma sebuah variabel yang berisi nilai yang sudah di asign di atas
 
        # Stop iteration if limit is reached
        if x > self.limit:
            raise StopIteration #ga ada keterangan sebuah eror kah kalo pakai StopIteration?
        #: StopIteration adalah exception khusus yang digunakan untuk menghentikan iterasi. Exception ini tidak memerlukan pesan error karena tujuannya adalah untuk memberi tahu loop bahwa iterasi sudah selesai, bukan untuk menunjukkan bahwa ada error atau masalah. Inilah yang menyebabkan loop for berhenti secara normal tanpa error.

        # Else increment and return old value
        self.x = x + 1;  # apakah code ini yang bikin method next jadi sebuah iterable return? 
        return x #knp bukan self.x?
 
# Prints numbers from 10 to 15
for i in Test(15):
    print(i)
 
# Prints nothing
for i in Test(5):
    print(i)
```
