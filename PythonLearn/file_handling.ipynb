## Impelemntasi operasi read



```python
# operasi secara normal dengan method read()
with open('handling.txt', 'r') as file:
    content = file.read()
    print(content)

```

```python
# operasi read dengan format perulangan setiap line. Bedanya hanya pada spasi antar line saja.
file = open('handling.txt', 'r')
for each in file:
    print(each)
```

```python
# implementasi read operasi dengan tujuan menampilkan spesifik beberapa karakter/huruf dalam baris saja.
file = open('handling.txt', 'r')
print(file.read(5))
```

```python
# implementasi read operasi dengan tujuan untuk menunjukan setiap lines by method readlines() lalu memisahkan setiap kata by split() method
with open("handling.txt", "r") as file:
    data = file.readlines()
    for line in data:
        word = line.split()
        print (word)
    
```

## code implementasi parameter seek() dan tell()

```python
# mengatur agar pinter ke index line pertama dan line spesifik dalam format standar
with open('example.txt', 'r') as file:
    # Membaca isi file
    content = file.read()
    print("Isi file pertama kali:", content)

    # file.seek(0)
    file.seek(10)

    
    # Membaca lagi dari awal
    content = file.read()
    print("Isi file kedua kali:", content)
```

```python
# contoh kasus:
# mengubah pointer dalam file.txt di suatu line
# inisiasi 
with open('example.txt', 'w') as file:
    file.write("0123456789abcdefghij")  # 20 karakter

# implementasi
with open('example.txt', 'r') as file:
    print("Posisi awal:", file.tell())
    
    # Mengatur pointer ke byte ke-10
    file.seek(10)
    print("Posisi setelah seek(10):", file.tell())
    
    # Membaca data dari file
    print("Data yang dibaca:", file.read(5)) #ketika ini dicoba running, maka ouput akan keluar dari index awal character 10 menuju 15 / 10+5
    print("Posisi setelah membaca 5 byte:", file.tell())
```

## implementasi penggunaan lstrip() method, rstrip() dan strip()

```python
with open('example.txt', 'w') as file:
    file.write("  Hello, World!  \n")
    file.write("  This is a test.  \n")
    file.write("  Leading and trailing spaces.  \n")

# Membaca file dan membersihkan whitespace
with open('example.txt', 'r') as file:
    for line in file:
        stripped_line_left = line.lstrip()  # Menghapus whitespace di awal
        stripped_line_right = line.rstrip() # Menghapus whitespace di akhir
        stripped_line_both = line.strip()   # Menghapus whitespace di awal dan akhir
        
        print(f"Asli: '{line}'")
        print(f"lstrip: '{stripped_line_left}'")
        print(f"rstrip: '{stripped_line_right}'")
        print(f"strip: '{stripped_line_both}'\n")
```

## impelemtnasi overall dalam bentuk method


```python
import os

def create_file(filename):
	try:
		with open(filename, 'w') as f:
			f.write('Hello, world!\n')
		print("File " + filename + " created successfully.")
	except IOError:
		print("Error: could not create file " + filename)

def read_file(filename):
	try:
		with open(filename, 'r') as f:
			contents = f.read()
			print(contents)
	except IOError:
		print("Error: could not read file " + filename)

def append_file(filename, text):
	try:
		with open(filename, 'a') as f:
			f.write(text)
		print("Text appended to file " + filename + " successfully.")
	except IOError:
		print("Error: could not append to file " + filename)

def rename_file(filename, new_filename):
	try:
		os.rename(filename, new_filename)
		print("File " + filename + " renamed to " + new_filename + " successfully.")
	except IOError:
		print("Error: could not rename file " + filename)

def delete_file(filename):
	try:
		os.remove(filename)
		print("File " + filename + " deleted successfully.")
	except IOError:
		print("Error: could not delete file " + filename)


if __name__ == '__main__':
	filename = "new.txt"
	new_filename = "new_example.txt"

	create_file(filename)
	read_file(filename)
	append_file(filename, "This is some additional text.\n")
	read_file(filename)
	rename_file(filename, new_filename)
	read_file(new_filename)
	delete_file(new_filename) #nyalakan bila perlu
```
