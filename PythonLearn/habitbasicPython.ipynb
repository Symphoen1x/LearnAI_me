# ilmu baru
* LOOPING:
* dalam looping ada yang namanya teknik penambahan (concatenation) string, itu adalah teknik yang setiap iterasi nya seperti ditambahkan satu persatu dari string awal yang digabung dengan string lain secara satu persatu.
* syarat dia sama kayak yang ada di while dimana define dahulu variabel diluar loop yang menjadi modal awal. misal suatu index value atau rentang value. pengkondisian bisa menjadi sebuah pointer/ref/tujuan dalam memproses sesuatu.
* contoh codenya ada di soal no 8. di soal itu, perulangan kerjanya itu kayak sesuai kondisi yang ada, jika dia sesuai yang pertama maka penampung akan menambahkan iterasi ke dalamnya. sama juga jika kondisi kedua, maka penampung juga menambahkan iterasi ke dalamnya. penampung harus diluar agar sistem bisa mudah membacanya.

Teknik yang digunakan dalam `bahan += i` adalah teknik penambahan (concatenation) string, di mana setiap iterasi, karakter `i` ditambahkan ke string `bahan`. Ini adalah cara untuk membangun atau memperbarui string secara bertahap.

Berikut adalah beberapa penggunaan `+=` untuk penambahan (concatenation) string dalam berbagai kondisi:

1. **Penambahan String dalam Loop**:
   - Digunakan untuk menggabungkan karakter satu per satu dari loop.
   ```python
   result = ""
   for char in "hello":
       result += char
   print(result)  # Output: hello
   ```

2. **Menyisipkan Karakter Khusus**:
   - Digunakan untuk menambahkan karakter khusus ke dalam string berdasarkan kondisi tertentu.
   ```python
   original = "hello"
   modified = ""
   for char in original:
       if char == 'e':
           modified += '3'
       else:
           modified += char
   print(modified)  # Output: h3llo
   ```

3. **Membangun String dari List**:
   - Digunakan untuk menggabungkan elemen dari sebuah list menjadi satu string.
   ```python
   words = ["hello", "world"]
   sentence = ""
   for word in words:
       sentence += word + " "
   print(sentence.strip())  # Output: hello world
   ```

4. **Menggabungkan String dengan Separator**:
   - Digunakan untuk menggabungkan string dengan pemisah tertentu.
   ```python
   words = ["apple", "banana", "cherry"]
   result = ""
   for word in words:
       result += word + ", "
   result = result.rstrip(", ")
   print(result)  # Output: apple, banana, cherry
   ```

5. **Menambahkan String pada Kondisi Tertentu**:
   - Digunakan untuk menambahkan string hanya jika kondisi tertentu terpenuhi.
   ```python
   numbers = [1, 2, 3, 4, 5]
   even_numbers = ""
   for number in numbers:
       if number % 2 == 0:
           even_numbers += str(number) + " "
   print(even_numbers.strip())  # Output: 2 4
   ```

6. **Menggabungkan String dalam Fungsi**:
   - Digunakan dalam fungsi untuk membangun string hasil secara iteratif.
   ```python
   def repeat_string(s, n):
       result = ""
       for _ in range(n):
           result += s
       return result
   print(repeat_string("abc", 3))  # Output: abcabcabc
   ```

**Penjelasan Kode yang Diberikan**:
Kode yang Anda berikan mencoba mengganti semua kemunculan karakter pertama dalam string dengan '$'. Berikut adalah versi yang disesuaikan agar bekerja dengan benar:

```python
user = str(input("masukan kata bebas: "))
s = user[1:len(user)]
bahan = user[0]
for i in s:
    if i == user[0]:
        bahan += "$"
    else:
        bahan += i
print(bahan)
```

Penjelasan:
- `user = str(input("masukan kata bebas: "))` mendapatkan input dari pengguna.
- `s = user[1:len(user)]` mengambil substring dari indeks ke-1 hingga akhir.
- `bahan = user[0]` menyimpan karakter pertama dari input.
- Loop `for i in s:` iterasi melalui setiap karakter dalam substring `s`.
- Jika `i == user[0]`, maka `i` diubah menjadi `"$"`.
- Karakter `i` atau `"$"` ditambahkan ke variabel `bahan` menggunakan `bahan += i`.
- String `bahan` dicetak setelah loop selesai.

Contoh input dan output:
- Input: `hello`
- Output: `h$llo`

Contoh input dan output:
- Input: `kata`
- Output: `k$ta`

````python
def change_char(str1):
    # Get the first character of the input string 'str1' and store it in the variable 'char'.
    char = str1[0]

    # Replace all occurrences of the character 'char' with '$' in the 'str1' string.
    str1 = str1.replace(char, '$')

    # Reconstruct the 'str1' string by placing the original 'char' as the first character
    # followed by the modified string starting from the second character.
    str1 = char + str1[1:]

    # Return the modified 'str1' string.
    return str1

# Call the change_char function with the argument 'restart' and print the result.
print(change_char('restart'))
````
