```python
import pandas as pd
df['Date'] = df['Date'].str.split(' → ').str[0] # Split the string on  ' → ' and keep only the first portion
df['Date'] = df['Date'].str.replace(' \(GMT\+7\)', '', regex=True) #Remove timezone info using regex
df['Date'] = pd.to_datetime(df['Date'], format='%B %d, %Y %I:%M %p') # Convert to datetime specifying the format
```

```python
import pandas as pd
df['Date'] = df['Date'].str.split(' → ').str[0] # Split the string on  ' → ' and keep only the first portion
df['Date'] = df['Date'].str.replace(' \(GMT\+7\)', '', regex=True) #Remove timezone info using regex
df['Date'] = pd.to_datetime(df['Date'], format='%B %d, %Y %H:%M') # Convert to datetime specifying the format, use %H for hours
```

```python
df.head(10)
```

```python
import matplotlib.pyplot as plt
```

```python
# prompt: berikan code untuk mengurutkan datatime pada kolom date dengan urutan paling lampau

df = df.sort_values(by='Date')
df.head(10)
```

```python
# Convert 'Yes'/'No' to binary (1/0) for all columns except 'Date', handling non-string values
def to_binary(val):
    if isinstance(val, str):
        return 1 if val.strip().lower() == 'yes' else 0
    return val
```

```python
# prompt: saya ingin apply the conversion, tapi jangan sampai ada kolom yang hilang

# Apply the conversion to all columns except 'Date'
for col in df.columns:
    if col != 'Date':
        df[col] = df[col].apply(to_binary)

df.head(10)
```

```python



# Calculate the sum of 'Yes' values (1s) for each row (corresponding to each date)
#Exclude the 'Date' column from the sum
df['count_yes'] = df.drop('Date', axis=1).sum(axis=1)


```

```python
import matplotlib.pyplot as plt

plt.figure(figsize=(10, 6))
plt.plot(df['Date'], df['count_yes'], marker='o')
plt.title('Scatter Plot of Yes Counts by Date with Lines')
plt.xlabel('Date')
plt.ylabel('Number of Yes Responses')

# Set x-axis ticks to match dates in df
plt.xticks(df['Date'], rotation=45)

plt.grid(True)
plt.show()
```
