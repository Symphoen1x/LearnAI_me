## mencoba bagaimana cara kerja method __call__
* kurang lebih kayak dibawah walaupun hasilnya bener-bener beda ama spacy. intinya 

```python
import spacy
# class yang menyimpan teknik pemecah
class Doc:
    def __init__(self, text): 
        self.text = text     
        self.tokens = [word for word in text.split(',')]
    def __iter__(self):
        for token in self.tokens:
            yield token

class Language:
    def __init__(self, model_name): #model_name dimasukan/inputan objek language
        self.model_name = model_name

    def __call__(self, text): #lalu class ini juga membuat inputan kedua sbg text yang nanti menjadi parameter dalam class Doc ke objeknya
        return Doc(text) #anggap aja kayak lu numpang nama disini, tapi udh keproses langsung gitu ke atas. Disinilah __call__ bekerja dengan membuat objek yang menampung class ini bakalan bisa input text. cek dibawah

# area pemanggilan
objekDoc = Language('en_core_web_sm')

# Process text to create a Doc object
doc = objekDoc("""bought trying""")

for word in doc:
    print(word)
```
