<h3>POS tags</h3>

```python
nlp = spacy.load("en_core_web_sm")
doc = nlp("Elon flew to mars yesterday. He carried biryani masala with him")

for token in doc:
    print(token," | ", token.pos_, " | ", spacy.explain(token.pos_))
```

<h3>Tags</h3>

```python
doc = nlp("Wow! Dr. Strange made 265 million $ on the very first day")

for token in doc:
    print(token," | ", token.pos_, " | ", spacy.explain(token.pos_), " | ", token.tag_, " | ", spacy.explain(token.tag_))
```

### Cara membedakan antara pos dengan tag
Dalam konteks *POS tagging* menggunakan spaCy, `pos_` dan `tag_` memang terlihat mirip, tetapi mereka sebenarnya memiliki perbedaan yang signifikan. Berikut adalah penjelasan perbedaan antara keduanya:

1. **POS (Part of Speech) - `pos_`:**
   - `pos_` mengacu pada kelas kata yang lebih umum, seperti kata benda (*noun*), kata kerja (*verb*), kata sifat (*adjective*), dll.
   - Ini memberikan informasi yang lebih umum dan biasanya merupakan kategori dasar dari kata tersebut.
   - Contoh kategori POS adalah: `NOUN`, `VERB`, `ADJ`, `ADV`, `PROPN`, dll.

2. **Tag - `tag_`:**
   - `tag_` memberikan informasi yang lebih spesifik mengenai bagian kata tersebut, biasanya dengan detail gramatikal tambahan.
   - `tag_` berfungsi sebagai label POS yang lebih terperinci, sering kali bergantung pada bahasa yang digunakan.
   - Contoh `tag_` untuk bahasa Inggris bisa termasuk: `NN` (noun singular), `NNS` (noun plural), `VBZ` (verb, third person singular present), `VBG` (verb, gerund or present participle), dll.
   - Jadi, meskipun dua kata mungkin memiliki `pos_` yang sama, mereka bisa memiliki `tag_` yang berbeda karena nuansa gramatikal.

**Contoh Penjelasan:**

Untuk kata "made" dalam kalimat di atas:
- **POS (`pos_`)**: `VERB` - Ini menunjukkan bahwa kata tersebut adalah kata kerja.
- **Tag (`tag_`)**: `VBD` - Ini menunjukkan bahwa kata kerja tersebut dalam bentuk lampau (past tense).

**Kesimpulan:**
- `pos_` memberikan informasi tentang kategori umum kata.
- `tag_` memberikan informasi lebih spesifik mengenai bentuk dan fungsi kata tersebut dalam kalimat.

Dengan memahami perbedaan ini, Anda bisa melihat bagaimana analisis linguistik bisa dilakukan pada tingkat yang lebih mendalam.

```python
earnings_text="""Microsoft Corp. today announced the following results for the quarter ended December 31, 2021, as compared to the corresponding period of last fiscal year:

·         Revenue was $51.7 billion and increased 20%
·         Operating income was $22.2 billion and increased 24%
·         Net income was $18.8 billion and increased 21%
·         Diluted earnings per share was $2.48 and increased 22%
“Digital technology is the most malleable resource at the world’s disposal to overcome constraints and reimagine everyday work and life,” said Satya Nadella, chairman and chief executive officer of Microsoft. “As tech as a percentage of global GDP continues to increase, we are innovating and investing across diverse and growing markets, with a common underlying technology stack and an operating model that reinforces a common strategy, culture, and sense of purpose.”
“Solid commercial execution, represented by strong bookings growth driven by long-term Azure commitments, increased Microsoft Cloud revenue to $22.1 billion, up 32% year over year” said Amy Hood, executive vice president and chief financial officer of Microsoft."""

doc = nlp(earnings_text)

filtered_tokens = []

for token in doc:
    if token.pos_ not in ["SPACE", "PUNCT", "X"]:
        filtered_tokens.append(token)
```
