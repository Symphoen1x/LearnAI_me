```python
#1. read the data provided in the same directory with name 'movies_sentiment_data.csv' and store it in df variable
df = pd.read_csv('IMDB Dataset.csv')

#2. print the shape of the data
print(df.shape)

#3. print top 5 datapoints
df.head()

```

```python
#check the distribution of 'Category' and see whether the Target labels are balanced or not.
df.sentiment.value_counts()
```

```python
#creating a new column "Category" which represent 1 if the sentiment is positive or 0 if it is negative
df['positive'] = df['sentiment'].apply(lambda x: 1 if x=='positive' else 0)

# def spam_or_not(x):
#     if x == 'positive':
#         return 1
#     else:
#         return 0

# df['positive'] = df['Category'].apply(spam_or_not)
df.head()
```

```python
#Do the 'train-test' splitting with test size of 20%
X_train, X_test, y_train, y_test = train_test_split(df.review, df.positive, test_size=0.2)
```

## Exercise-1

1. using sklearn pipeline module create a classification pipeline to classify the movie review's positive or negative.

**Note:**
- use CountVectorizer for pre-processing the text.

- use **Random Forest** as the classifier with estimators as 50 and criterion as entropy.
- print the classification report.

**References**:

- https://scikit-learn.org/stable/modules/generated/sklearn.ensemble.RandomForestClassifier.html

- https://scikit-learn.org/stable/modules/generated/sklearn.feature_extraction.text.CountVectorizer.html

```python
# Create bag of words representation using CountVectorizer
cv = CountVectorizer()
fit_train = cv.fit_transform(X_train.values)
# fit_train.toarray() #ketika proses mengubah ke array(readable) ini di run kemungkinan ke riset runtimenya karena gede banget. so, fokus ke tujuan exercise!
# list_feature = cv.get_feature_names_out()
# print(list_feature)


# Train the Random Forest model
model = RandomForestClassifier(n_estimators=50, criterion='entropy')


# fit with X_train and y_train
model.fit(fit_train, y_train)

# x_test
x_test_cv = cv.transform(X_test)

# get the predictions for X_test and store it in y_pred
y_pred = model.predict(cv.transform(x_test_cv))


# print the classfication report
print(classification_report(y_test, y_pred))




# create a pipeline object
clf = Pipeline([
    ('vectorizer', cv),
    ('random_forest', model)
])
print(clf)
```
