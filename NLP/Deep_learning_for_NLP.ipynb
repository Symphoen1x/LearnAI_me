* Alasan mengapa membutuhkan Activation function di neural network
<br>
  Alasanya lebih ke arah:<br>
  1. Activation function sangat membantu dalam model mencapai target untuk suatu tugas seperti classifikasi, regresi, dll berdasarkan persebaran pola data.<br>
  2. Jika tugasnya regresi yang berhubungan dengan garis linear dalam menentukan sesuatu, garis tersebut bisa dicapai dengan memanfaatkan perhitungan input dalam suatu fungsi linear yang diusahakan mendekati garis/target yang akan dicapai. <br>
  3. Sementara, tugas lain seperti clasifikasi walaupun ini bisa membentuk pola data linear dalam target/garis yang ingin didekati, ada juga pola persebaran yang bentuknya non-linear sehingga membutuhkan fungsi aktivasi lain khusus non-linear.
* Activation function berdasarkan jenis neuron di suatu layer
<br>
 1.  Ada dua jenis layer yang ada di struktur ANN yang membutuhkan aktifasi fungsi, yakni hidden layer dan output layer.
 2. Dari sumber PDF, beberapa fungsi aktifasi yang sering/secara history digunakan di hidden layer seperti linear aktifasi, sigmoid, ReLU, step function, dll.
 3. sementara, fugnsi aktifasi yang biasanya khusus diterapkan pada output layer/neuron di dalamnya seperti softmax, dll.
* Penjelasan mendalam, tapi inti terkait beberapa aktifasi fungsi. Pertama: <br>
  1. Step Activation Function. <br>
  * Fungsi ini sudah cukup jarang digunakan karena keterbatasanya dalam menangkap pola persebaran data. Dari segi pola fungsinya, dia mungkin lebih cocok untuk tugas yang persebaran datanya linear.
  * Intinya, dia punya rules jika hasil rumus w.x + b adalah lebih besar daripada 0, hasilnya bernilai 1. Sementara jika kurang dari dan sama dengan 0, hasilnya bernilai pasti 0. Ini tentu menyulitkan beberapa tugas meskipun linearitas bentuk polanya.
  * Fungsi ini tidak bersifat bijektif(setiap input memiliki output yg unik dan sebaliknya) dan tidak injektif/surjektif(satu ke satu) karena input yang berbeda bisa saja menghasilkan output yang sama seperti x = 0.5 dan x=0.9 hasilnya sama yaitu 1. Begitupun input negatif yg berbeda.
  2. Linear Activation function.<br>
  Dari namanya, sudah tidak asing bahwa dia adalah fungsi khusus untuk tugas Linearitas. Berbeda dnegan funsi sebelumnya, fungsi ini linear, tapi rulesnya y=x. artinya apabila kita membagi nilai y(input)/x(input), hasilnya = 1. Sangat cocok untuk tugas regresi dan klasifikasi.
  3. Sigmoid activation function <br>
  * Fungsi yang juga mengisi gap step funct karena lebih granual(detail dan jelas) juga informatif.
  * Rulesnya bisa cek di image.
  * Fungsi ini memotong sumbu y di titik 0,5 karena bila input(x) yang dimasukan adalah 0 maka hasilnya/outputnya adalah pembagian 1/2 atau 0,5.
  * Jika x = 0, y=0.5, input dibawah 0/nefatif tak hingga akan dikonversi menjadi 0. Atau bisa dibilang input < 0 = 0. Sebaliknya, input di atas 0, akan menyesuaikan hasil dari pembagian sesuai rumus. Rangenya hanya 0-1.
  4. The Rectified Linear Activation Function
  * Mirip seperti Linear function, hanya beda di rumusnya.
  * Jika ingin memahami proses kerja fugnsi ini terutama pada grafik carilah contoh gambaran yang prosesnya ada perhitungan rumus NN + inputnya dan ada output. Biar lebih mudah dan cepat paham
  5. Softmax activation function
  * fungsi yang memanfaatkan euler's number yang merupakan constanta exponen untuk proses mengubah elemen dalam index vektor menjadi decimal range 0-1 karena hasil normalisasi pembagian eksponensial tunggal/penjumlahan eksponensial.
  *  Jika softmax tidak menggunakan eksponensial, nilai input yang sangat besar atau kecil dapat mendominasi perhitungan, membuat distribusi probabilitas menjadi tidak stabil. Fungsi eksponensial memperkuat perbedaan antara nilai input, yang membantu memperjelas urutan probabilitas, dan mengurangi dampak nilai yang sangat besar atau sangat kecil.
  * mungin terkesan mustahil bisa dalam range seperti itu, tapi silahkan dibuktikan sendiri

## Gambaran kenapa neural network dengan fungsi aktivasi nonlinear dapat menangani data nonlinear dengan baik

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.neural_network import MLPClassifier
from sklearn.datasets import make_moons
from sklearn.model_selection import train_test_split

# 1. Membuat dataset dua dimensi yang sulit dipisahkan secara linear
X, y = make_moons(n_samples=300, noise=0.2, random_state=42)
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)

# 2. Membuat model neural network dengan hidden layers
model = MLPClassifier(hidden_layer_sizes=(10, 10), activation='relu', max_iter=1000, random_state=42)
model.fit(X_train, y_train)

# 3. Visualisasi garis batas keputusan
# Membuat grid untuk setiap titik di area plot
x_min, x_max = X[:, 0].min() - 0.5, X[:, 0].max() + 0.5
y_min, y_max = X[:, 1].min() - 0.5, X[:, 1].max() + 0.5
xx, yy = np.meshgrid(np.arange(x_min, x_max, 0.01), np.arange(y_min, y_max, 0.01))

# Memprediksi label untuk setiap titik di grid
Z = model.predict(np.c_[xx.ravel(), yy.ravel()])
Z = Z.reshape(xx.shape)

# Plot area klasifikasi dan titik data
plt.contourf(xx, yy, Z, alpha=0.3, cmap=plt.cm.Spectral)
plt.scatter(X_train[:, 0], X_train[:, 1], c=y_train, marker='o', s=30, cmap=plt.cm.Spectral, edgecolor='k', label="Training data")
plt.scatter(X_test[:, 0], X_test[:, 1], c=y_test, marker='s', s=30, cmap=plt.cm.Spectral, edgecolor='k', label="Test data")
plt.title("Garis Batas Keputusan Neural Network")
plt.xlabel("Fitur 1")
plt.ylabel("Fitur 2")
plt.legend()
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.neural_network import MLPClassifier
from sklearn.datasets import make_moons
from sklearn.model_selection import train_test_split

# Membuat dataset yang tidak bisa dipisahkan secara linear
X, y = make_moons(n_samples=300, noise=0.2, random_state=42)
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)

# Model 1: Neural Network dengan fungsi aktivasi non-linear (ReLU)
model_non_linear = MLPClassifier(hidden_layer_sizes=(10, 10), activation='relu', max_iter=1000, random_state=42)
model_non_linear.fit(X_train, y_train)

# Model 2: Neural Network tanpa fungsi aktivasi non-linear (fungsi aktivasi 'identity' yang linear)
model_linear = MLPClassifier(hidden_layer_sizes=(10, 10), activation='identity', max_iter=1000, random_state=42)
model_linear.fit(X_train, y_train)

# Fungsi untuk plot garis batas keputusan
def plot_decision_boundary(model, X, y, title):
    x_min, x_max = X[:, 0].min() - 0.5, X[:, 0].max() + 0.5
    y_min, y_max = X[:, 1].min() - 0.5, X[:, 1].max() + 0.5
    xx, yy = np.meshgrid(np.arange(x_min, x_max, 0.01), np.arange(y_min, y_max, 0.01))

    Z = model.predict(np.c_[xx.ravel(), yy.ravel()])
    Z = Z.reshape(xx.shape)

    plt.contourf(xx, yy, Z, alpha=0.3, cmap=plt.cm.Spectral)
    plt.scatter(X[:, 0], X[:, 1], c=y, marker='o', s=30, cmap=plt.cm.Spectral, edgecolor='k')
    plt.title(title)
    plt.xlabel("Fitur 1")
    plt.ylabel("Fitur 2")
    plt.show()

# Plot garis batas keputusan untuk model non-linear
plot_decision_boundary(model_non_linear, X, y, "Neural Network dengan Aktivasi Non-Linear (ReLU)")

# Plot garis batas keputusan untuk model linear
plot_decision_boundary(model_linear, X, y, "Neural Network dengan Aktivasi Linear (Tanpa Non-Linearitas)")
```
