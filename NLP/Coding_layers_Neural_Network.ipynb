
#### USING LOOPS FOR BETTER AND EASIER CODING

```python
inputs = [1, 2, 3, 2.5]

##LIST OF 3 WEIGHTS
weights = [[0.2, 0.8, -0.5, 1],
 [0.5, -0.91, 0.26, -0.5],
 [-0.26, -0.27, 0.17, 0.87]]

##LIST OF BIASES
biases = [2, 3, 0.5]

# Output of current layer
layer_outputs = []

# For each neuron
for neuron_weights, neuron_bias in zip(weights, biases):
 # Zeroed output of given neuron
 neuron_output = 0
 # For each input and weight to the neuron
 for n_input, weight in zip(inputs, neuron_weights):
 # Multiply this input by associated weight
 # and add to the neuron's output variable
   neuron_output += n_input*weight ## W31*X1 + W32*X2 + W33*X3 + W34*X4
   # Add bias
 neuron_output += neuron_bias ## ## W31*X1 + W32*X2 + W33*X3 + W34*X4 + B3
 # Put neuron's result to the layer's output list
 layer_outputs.append(neuron_output)
print(layer_outputs)
```

```python
# Dense layer
class Layer_Dense:
 # Layer initialization
 def __init__(self, n_inputs, n_neurons):
 # Initialize weights and biases
   self.weights = 0.01 * np.random.randn(n_inputs, n_neurons)
   self.biases = np.zeros((1, n_neurons))
 # Forward pass
 def forward(self, inputs):
 # Calculate output values from inputs, weights and biases
   self.output = np.dot(inputs, self.weights) + self.biases
```

```python
# Softmax activation
class Activation_Softmax:
 # Forward pass
 def forward(self, inputs):
 # Get unnormalized probabilities
  exp_values = np.exp(inputs - np.max(inputs, axis=1, keepdims=True))
 # Normalize them for each sample
  probabilities = exp_values / np.sum(exp_values, axis=1,keepdims=True)
  self.output = probabilities
```

#### IMPLEMENTING THE LOSS CLASS

```python
import numpy as np
```

```python
# konteks: class loss guna menghasilkan nilai rata-rata loss diseluruh sampel yang telah diproses dengan cross-entropy
class Loss:
 # Calculates the data and regularization losses
 # given model output and ground truth values
 def calculate(self, output, y):
  # Calculate sample losses
  sample_losses = self.forward(output, y)
  # Calculate mean loss
  data_loss = np.mean(sample_losses)
  # Return loss
  return data_loss
```

#### IMPLEMENTING THE CATEGORICAL CROSS ENTROPY CLASS


```python
# konteks: ketika cross-entropy disatukan dengan loss menggunakan class turunan (loss -> cross-entropy)
# Cross-entropy loss
class Loss_CategoricalCrossentropy(Loss):
 # Forward pass
 def forward(self, y_pred, y_true):
  # Number of samples in a batch
  samples = len(y_pred)
  # Clip data to prevent division by 0
  # Clip both sides to not drag mean towards any value
  y_pred_clipped = np.clip(y_pred, 1e-7, 1 - 1e-7)
  # Probabilities for target values -
  # only if categorical labels
  if len(y_true.shape) == 1:
   correct_confidences = y_pred_clipped[
   range(samples),
   y_true
   ]
  # Mask values - only for one-hot encoded labels
  elif len(y_true.shape) == 2:
   correct_confidences = np.sum(
   y_pred_clipped*y_true,
   axis=1
   )
 # Losses
  negative_log_likelihoods = -np.log(correct_confidences)
  return negative_log_likelihoods
```

```python
# hasil final loss function x cross-entropy
softmax_outputs = np.array([[0.7, 0.1, 0.2],
 [0.1, 0.5, 0.4],
 [0.02, 0.9, 0.08]])
class_targets = np.array([[1, 0, 0],
 [0, 1, 0],
 [0, 1, 0]])
loss_function = Loss_CategoricalCrossentropy()
loss = loss_function.calculate(softmax_outputs, class_targets)
print(loss)
```

#### The slope or approximate_derivative value of tangent line with small delta value

```python
# Bukti kalo misal tanpa small delta dan input (x) tidak ada perubahan, tapi hasil fungsi ada
def hitung_slope(x1, y1, x2, y2):
    """
    Menghitung slope (kemiringan garis) antara dua titik.
    Returns:
    - float: nilai slope jika valid
    - string: pesan error jika delta x = 0
    """
    try:
        delta_x = x2 - x1
        delta_y = y2 - y1

        # Cek apakah delta_x adalah 0
        if delta_x == 0:
            raise ZeroDivisionError("Delta x = 0: Garis vertikal, slope tidak terdefinisi")

        slope = delta_y / delta_x
        return slope

    except ZeroDivisionError as e:
        return str(e)

# Demonstrasi kasus-kasus
print("Kasus 1: Garis normal")
print(hitung_slope(1, 1, 2, 2))  # Seharusnya menghasilkan slope = 1

print("\nKasus 2: Garis vertikal (delta_x = 0)")
print(hitung_slope(3, 1, 3, 5))  # Seharusnya menghasilkan error

print("\nKasus 3: Garis horizontal (delta_y = 0)")
print(hitung_slope(1, 3, 4, 3))  # Seharusnya menghasilkan slope = 0
```

```python
class Activation_Softmax:

    # Forward pass
    def forward(self, inputs):
        # Remember input values
        self.inputs = inputs

        # Get unnormalized probabilities
        exp_values = np.exp(inputs - np.max(inputs, axis=1,
                                            keepdims=True))
        # Normalize them for each sample
        probabilities = exp_values / np.sum(exp_values, axis=1,
                                            keepdims=True)

        self.output = probabilities

    # Backward pass
    def backward(self, dvalues):

        # Create uninitialized array
        self.dinputs = np.empty_like(dvalues)
        print(self.dinputs)

        # Enumerate outputs as vector and gradients as vector
        for index, (single_output, single_dvalues) in enumerate(zip(self.output, dvalues)):
            # Flatten output array (awal 1 d setiap value divektor, skrng jadi 2d)
            single_output = single_output.reshape(-1, 1)
            print(single_output)
            # Calculate Jacobian matrix of the output and
            # 3x3
            # dikurangi
            # 3*1 * 1*3 = 3*3
            jacobian_matrix = np.diagflat(single_output) - np.dot(single_output, single_output.T)
            print(jacobian_matrix)
            # Calculate sample-wise gradient
            # and add it to the array of sample gradients
            # ini bagian derivative
            self.dinputs[index] = np.dot(jacobian_matrix,
                                         single_dvalues)
```

```python
class Activation_Softmax:

    # Forward pass
    def forward(self, inputs):
        # Remember input values
        self.inputs = inputs

        # Get unnormalized probabilities
        exp_values = np.exp(inputs - np.max(inputs, axis=1,
                                            keepdims=True))
        # Normalize them for each sample
        probabilities = exp_values / np.sum(exp_values, axis=1,
                                            keepdims=True)

        self.output = probabilities

    # Backward pass
    def backward(self, dvalues):

        # Create uninitialized array
        self.dinputs = np.empty_like(dvalues)

        # Enumerate outputs as vector and gradients as vector
        for index, (single_output, single_dvalues) in enumerate(zip(self.output, dvalues)):
            # Flatten output array (awal 1 d setiap value divektor, skrng jadi 2d)
            single_output = single_output.reshape(-1, 1)
            # print(single_output)
            # Calculate Jacobian matrix of the output and
            # 3x3
            # dikurangi
            # 3*1 * 1*3 = 3*3
            jacobian_matrix = np.diagflat(single_output) - np.dot(single_output, single_output.T)
            print(f'der softmaxt to output: {jacobian_matrix}')
            # Calculate sample-wise gradient
            # and add it to the array of sample gradients
            # ini bagian derivative
            self.dinputs[index] = np.dot(jacobian_matrix,
                                         single_dvalues)
```

```python
class Loss_CategoricalCrossentropy():
  def forward(self, y_pred, y_true):
      # Number of samples in a batch
      samples = len(y_pred)
      # Clip data to prevent division by 0
      # Clip both sides to not drag mean towards any value
      y_pred_clipped = np.clip(y_pred, 1e-7, 1 - 1e-7)
      # Probabilities for target values -
      # only if categorical labels
      if len(y_true.shape) == 1:
          correct_confidences = y_pred_clipped[
          range(samples),
          y_true
          ]
      # Mask values - only for one-hot encoded labels
      elif len(y_true.shape) == 2:
          correct_confidences = np.sum(
          y_pred_clipped*y_true,
          axis=1
          )
 #     Losses
      negative_log_likelihoods = -np.log(correct_confidences)
      return negative_log_likelihoods
    # Backward pass
  def backward(self, dvalues, y_true ):
        # Number of samples
        samples = len(dvalues)
        print(f'check ukuran samples dari softmax output: {samples}')
        # Number of labels in every sample
        # We'll use the first sample to count them
        labels = len(dvalues[0])
        # If labels are sparse, turn them into one-hot vector
        if len(y_true.shape) == 1:
            y_true = np.eye(labels)[y_true]
            print(f'Jika label adalah vector: {y_true}')

        # Calculate gradient
        self.dinputs = -y_true / dvalues
        # Normalize gradient
        self.dinputs = self.dinputs / samples
```

```python
# training begin
class trainingManual:
  def __init__(self, totalIteration, adjustW, adjustB):
    self.totalIteration = totalIteration
    self.adjustW = adjustW
    self.adjustB = adjustB

  def training(self)
    for iteration in range(self.totalIteration):
        # Update weights with some small random values
        dense1.weights += self.adjustW * np.random.randn(2, 3)
        dense1.biases +=  self.adjustB * np.random.randn(1, 3)
        dense2.weights += self.adjustW * np.random.randn(3, 3)
        dense2.biases += self.adjustB * np.random.randn(1, 3)

        # Perform a forward pass of our training data through this layer
        dense1.forward(X)
        activation1.forward(dense1.output)
        dense2.forward(activation1.output)
        activation2.forward(dense2.output)

        # Perform a forward pass through activation function
        # it takes the output of second dense layer here and returns loss
        loss = loss_function.calculate(activation2.output, y)
        # Calculate accuracy from output of activation2 and targets
        # calculate values along first axis
        predictions = np.argmax(activation2.output, axis​=1)
        accuracy = np.mean(predictions==y)
        # If loss is smaller - print and save weights and biases aside
        if loss < lowest_loss:
          print('New set of weights found, iteration:', iteration,
                    'loss:', loss, 'acc:', accuracy)
          best_dense1_weights = dense1.weights.copy()
          best_dense1_biases = dense1.biases.copy()
          best_dense2_weights = dense2.weights.copy()
          best_dense2_biases = dense2.biases.copy()
          lowest_loss = loss
        # Revert weights and biases
        else:
          dense1.weights = best_dense1_weights.copy()
          dense1.biases = best_dense1_biases.copy()
          dense2.weights = best_dense2_weights.copy()
          dense2.biases = best_dense2_biases.copy()
```
