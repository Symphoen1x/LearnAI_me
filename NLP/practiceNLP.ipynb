## penjelasan code dibalik PorterStemmer
Tentu, saya bisa memberikan gambaran tentang bagaimana `PorterStemmer` bekerja. Algoritma Porter Stemmer adalah algoritma yang populer dan banyak digunakan untuk stemming, dan kode aslinya tersedia dalam beberapa bahasa pemrograman. Berikut adalah implementasi sederhana dari algoritma Porter Stemmer dalam Python:

```python
class PorterStemmer:
    def __init__(self):
        self.vowels = "aeiou"
        self.double_consonants = ["bb", "dd", "ff", "gg", "mm", "nn", "pp", "rr", "tt"]
        self.li_ending = "cdeghkmnrt"
        self.step_2_suffixes = {
            "ational": "ate", "tional": "tion", "enci": "ence", "anci": "ance",
            "izer": "ize", "bli": "ble", "alli": "al", "entli": "ent",
            "eli": "e", "ousli": "ous", "ization": "ize", "ation": "ate",
            "ator": "ate", "alism": "al", "iveness": "ive", "fulness": "ful",
            "ousness": "ous", "aliti": "al", "iviti": "ive", "biliti": "ble",
            "logi": "log"
        }
        self.step_3_suffixes = {
            "icate": "ic", "ative": "", "alize": "al", "iciti": "ic",
            "ical": "ic", "ful": "", "ness": ""
        }

    def is_consonant(self, word, i):
        """Check if the letter at index i is a consonant."""
        if word[i] in self.vowels:
            return False
        if word[i] == 'y':
            if i == 0:
                return True
            else:
                return not self.is_consonant(word, i - 1)
        return True

    def measure(self, word):
        """Measure the number of consonant sequences (measures)."""
        cv_seq = []
        for i in range(len(word)):
            if self.is_consonant(word, i):
                if not cv_seq or cv_seq[-1] != 'C':
                    cv_seq.append('C')
            else:
                if not cv_seq or cv_seq[-1] != 'V':
                    cv_seq.append('V')
        return cv_seq.count('VC')

    def contains_vowel(self, stem):
        """Check if the stem contains a vowel."""
        for letter in stem:
            if letter in self.vowels:
                return True
        return False

    def ends_double_consonant(self, word):
        """Check if the word ends with a double consonant."""
        return len(word) >= 2 and word[-2:] in self.double_consonants

    def ends_cvc(self, word):
        """Check if the word ends with consonant-vowel-consonant and the final consonant is not w, x, or y."""
        if len(word) >= 3 and self.is_consonant(word, -1) and not self.is_consonant(word, -2) and self.is_consonant(word, -3):
            if word[-1] not in "wxy":
                return True
        return False

    def step_1a(self, word):
        if word.endswith("sses"):
            return word[:-2]
        if word.endswith("ies"):
            return word[:-2]
        if word.endswith("ss"):
            return word
        if word.endswith("s"):
            return word[:-1]
        return word

    def step_1b(self, word):
        if word.endswith("eed"):
            stem = word[:-3]
            if self.measure(stem) > 0:
                return stem + "ee"
            return word
        if word.endswith("ed"):
            stem = word[:-2]
            if self.contains_vowel(stem):
                return self.step_1b2(stem)
            return word
        if word.endswith("ing"):
            stem = word[:-3]
            if self.contains_vowel(stem):
                return self.step_1b2(stem)
            return word
        return word

    def step_1b2(self, word):
        if word.endswith("at") or word.endswith("bl") or word.endswith("iz"):
            return word + "e"
        if self.ends_double_consonant(word):
            return word[:-1]
        if self.measure(word) == 1 and self.ends_cvc(word):
            return word + "e"
        return word

    def step_1c(self, word):
        if word.endswith("y"):
            stem = word[:-1]
            if self.contains_vowel(stem):
                return stem + "i"
        return word

    def step_2(self, word):
        for suffix, replacement in self.step_2_suffixes.items():
            if word.endswith(suffix):
                stem = word[:-len(suffix)]
                if self.measure(stem) > 0:
                    return stem + replacement
        return word

    def step_3(self, word):
        for suffix, replacement in self.step_3_suffixes.items():
            if word.endswith(suffix):
                stem = word[:-len(suffix)]
                if self.measure(stem) > 0:
                    return stem + replacement
        return word

    def step_4(self, word):
        suffixes = ["al", "ance", "ence", "er", "ic", "able", "ible", "ant", "ement", "ment", "ent", "ou", "ism", "ate", "iti", "ous", "ive", "ize"]
        for suffix in suffixes:
            if word.endswith(suffix):
                stem = word[:-len(suffix)]
                if self.measure(stem) > 1:
                    return stem
        if word.endswith("ion"):
            stem = word[:-3]
            if self.measure(stem) > 1 and (stem.endswith("s") or stem.endswith("t")):
                return stem
        return word

    def step_5a(self, word):
        if word.endswith("e"):
            stem = word[:-1]
            if self.measure(stem) > 1 or (self.measure(stem) == 1 and not self.ends_cvc(stem)):
                return stem
        return word

    def step_5b(self, word):
        if self.measure(word) > 1 and self.ends_double_consonant(word) and word.endswith("l"):
            return word[:-1]
        return word

    def stem(self, word):
        word = word.lower()
        word = self.step_1a(word)
        word = self.step_1b(word)
        word = self.step_1c(word)
        word = self.step_2(word)
        word = self.step_3(word)
        word = self.step_4(word)
        word = self.step_5a(word)
        word = self.step_5b(word)
        return word

# Contoh penggunaan
stemmer = PorterStemmer()
words = ["easily", "running", "better"]
stemmed_words = [stemmer.stem(word) for word in words]

print(stemmed_words)  # Output: ['easili', 'run', 'better']
```

**Penjelasan Kode**:

1. **Konstruktor (`__init__` method):**
   - Inisialisasi beberapa atribut yang akan digunakan dalam algoritma, seperti daftar vokal, konsonan ganda, dan beberapa aturan untuk tahap-tahap tertentu dalam stemming.

2. **Metode `is_consonant`:**
   - Memeriksa apakah karakter pada posisi tertentu dalam kata adalah konsonan.

3. **Metode `measure`:**
   - Menghitung jumlah pola konsonan-vokal dalam kata, yang disebut sebagai "measures".

4. **Metode `contains_vowel`:**
   - Memeriksa apakah kata (stem) mengandung vokal.

5. **Metode `ends_double_consonant`:**
   - Memeriksa apakah kata berakhir dengan dua konsonan yang sama.

6. **Metode `ends_cvc`:**
   - Memeriksa apakah kata berakhir dengan pola konsonan-vokal-konsonan, dengan pengecualian untuk huruf "w", "x", atau "y".

7. **Langkah 1a hingga 1c:**
   - Menghapus akhiran yang umum seperti "sses", "ies", "ss", dan "s", serta menangani akhiran "ed" dan "ing".

8. **Langkah 2 dan 3:**
   - Mengganti akhiran tertentu berdasarkan aturan yang telah ditentukan jika measure cukup besar.

9. **Langkah 4 dan 5:**
   - Menghapus atau mengganti akhiran berdasarkan kondisi yang lebih kompleks.

10. **Metode `stem`:**
    - Menggabungkan semua langkah untuk mengurangi kata menjadi bentuk dasarnya.

Ini adalah gambaran dasar dari algoritma Porter Stemmer. Implementasi aslinya bisa lebih kompleks dan memiliki lebih banyak rincian, tetapi contoh ini memberikan ide dasar tentang bagaimana algoritma tersebut bekerja.
