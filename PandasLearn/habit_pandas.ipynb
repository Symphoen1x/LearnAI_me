## 11. Write a Pandas program to select the rows where number of attempts in the examination is less than 2 and score greater than 15.

Sample  Python dictionary data and list labels:
exam_data = {'name': ['Anastasia', 'Dima', 'Katherine', 'James', 'Emily', 'Michael', 'Matthew', 'Laura', 'Kevin', 'Jonas'],
'score': [12.5, 9, 16.5, np.nan, 9, 20, 14.5, np.nan, 8, 19],

```python
Expected Output:
Number of attempts in the examination is less than 2 and score greater than 15 :
    name  score  attempts qualify
j  Jonas   19.0         1     yes
```

```python
import pandas as pd
import numpy as np
exam_data = {'name': ['Anastasia', 'Dima', 'Katherine', 'James', 'Emily', 'Michael', 'Matthew', 'Laura', 'Kevin', 'Jonas'],
'score': [12.5, 9, 16.5, np.nan, 9, 20, 14.5, np.nan, 8, 19],
'attempts': [1, 3, 2, 3, 2, 3, 1, 1, 2, 1],
'qualify': ['yes', 'no', 'yes', 'no', 'no', 'yes', 'yes', 'no', 'no', 'yes']}
labels = ['a', 'b', 'c', 'd', 'e', 'f', 'g', 'h', 'i', 'j']
# df = pd.DataFrame(columns= ['attempts', 'name', 'qualify', 'score'], data = exam_data, index=labels)
df = pd.DataFrame(exam_data, index = labels)
df = df[(df['attempts'] < 2) & (df['score'] > 15)] #sementara ini adalah operator beteween wise yang dalam matematika kayak 1 < x < 5 disederhanakan menjadi A > 1 & A < 5 dan bisa untuk variabel yang berbeda A < 1 & B > 2. Operator ini hubunganya sama perkalian
# df = df[(df['attempts'] < 2) and (df['score'] > 15)] #bedakan ini dengan yang diatas, untuk and sendiri adalah ekspresi kedua variabel yand dipilih pada konteks logika boolean. satu variabel punya value (true/false) dan itu diapkai buat pengkondisian

df
```
