````python
# series/vector operation
import pandas as pd
data = [10, 20, 30]
jumlah = pd.Series(data, index = ["a", "b", "c"])
# penjumlahan elemen
jumlah += 5
print(jumlah)

kurang = pd.Series(data, index = ["a", "b", "c"])
kurang -= 5
print(kurang)



kali = pd.Series(data, index = ["a", "b", "c"])
kali *= 5
print(kali)


bagi = pd.Series(data, index = ["a", "b", "c"])
bagi /= 5
print(bagi)



````

````python
# statistik series
data = [10, 20, 30]
series = pd.Series(data, index = ["a", "b", "c"])
# Rata-rata nilai Series
mean_value = series.mean()
print("Rata-rata nilai Series:", mean_value)

# Jumlah total nilai Series
sum_value = series.sum()
print("Jumlah total nilai Series:", sum_value)

# Nilai maksimum dari Series
max_value = series.max()
print("Nilai maksimum Series:", max_value)

# Nilai minimum dari Series
min_value = series.min()
print("Nilai minimum Series:", min_value)
````

### Reindexing
---
`reindex()` adalah method yang berfungsi untuk:

- **Perubahan Index setelah Filtering dan Subsetting**:
  - Setelah melakukan filter atau subsetting pada DataFrame, indeks mungkin tidak lagi berurutan. `reindex()` membantu memastikan indeks berurutan dan dimulai dari nol.

- **Membantu Mengisi dan Mengganti Missing Value**:
  - Dengan menambahkan kolom baru yang berisi indeks yang lengkap, `reindex()` membantu mengisi nilai yang hilang dengan nilai default (biasanya NaN) atau dengan nilai yang diinginkan.

- **Perubahan Urutan Index untuk Analisis Visual**:
  - Dalam analisis visual data tertentu, di mana nilai atau baris yang sejajar dengan indeks memiliki banyak arti, `reindex()` membantu mengatur ulang indeks agar lebih mudah dilihat dan diinterpretasikan.

- **Mempermudah Teknik Join pada Merge Case Dataset**:
  - Saat menggabungkan dua DataFrame dengan `join` atau `merge`, `reindex()` membantu menyesuaikan indeks untuk memastikan data yang tepat sesuai dengan indeks yang diinginkan.
```

Manfaat dalam Penanganan Data
- **Konsistensi Data**: Memastikan konsistensi indeks dalam DataFrame.
- **Kelengkapan Data**: Membantu dalam pengisian nilai yang hilang dan menambah kolom baru.
- **Kemudahan Visualisasi**: Mengatur ulang indeks agar lebih mudah diinterpretasikan dalam analisis visual.
- **Efisiensi dalam Penggabungan Data**: Menyesuaikan indeks untuk mempermudah proses join pada merge case dataset.

Kesimpulan
Metode `reindex()` pada pandas adalah alat yang kuat dalam penanganan data, membantu dalam berbagai situasi seperti pengolahan data yang konsisten, pengisian nilai yang hilang, penyusunan ulang data untuk analisis visual, dan menyederhanakan proses penggabungan data dari berbagai sumber.

### About Empty cell

````python
# conoth code pembuatan parameter yang bisa didefine beserta valuenya -> inplcae = True
class MyDataFrame:
    def __init__(self, data):
        self.data = data

    def drop_na(self, inplace=False):
        if inplace:
            # Modify the original data
            self.data = [value for value in self.data if value is not None]
        else:
            # Return a new instance with missing values removed
            new_data = [value for value in self.data if value is not None]
            return MyDataFrame(new_data)

# Contoh penggunaan
df = MyDataFrame([1, None, 2, None, 3])

# Menggunakan inplace=False (default)
new_df = df.drop_na()
print("Original data:", df.data)  # Output: Original data: [1, None, 2, None, 3]
print("New data:", new_df.data)   # Output: New data: [1, 2, 3]

# Menggunakan inplace=True
df.drop_na(inplace=True)
print("Modified original data:", df.data)  # Output: Modified original data: [1, 2, 3]
````

## Pembahasan lebih lanjut terkait method korelasi corr()
1. Korelasi Pearson
* Tujuan: Mengukur hubungan linier antara dua variabel.
* Rumus:
covariance(X, Y)/standar devisiasi X . standar devisiasi Y

Koefisien korelasi Pearson r dihitung sebagai rasio antara kovarian dari dua variabel dan hasil kali dari deviasi standar masing-masing variabel.
* Nilai: Bervariasi antara -1 dan 1.
* Kondisi Penggunaan yang Cocok:
Data Linier: Digunakan ketika ada hubungan linier yang kuat antara dua variabel.
Distribusi Normal: Paling cocok jika data berdistribusi normal.
Sensitif terhadap Outliers: Karena outliers dapat mempengaruhi hasil secara signifikan, sebaiknya digunakan pada data tanpa outliers atau data yang telah dibersihkan dari outliers.
2. Korelasi Kendall (Kendall Tau)
* Tujuan: Mengukur hubungan monotonic antara dua variabel menggunakan peringkat.
* Rumus: Berdasarkan jumlah pasangan data yang memiliki urutan yang sama atau berlawanan.
* Nilai: Bervariasi antara -1 dan 1.
* Kondisi Penggunaan yang Cocok:
Data Berperingkat: Cocok untuk data ordinal atau data yang berperingkat.
Hubungan Non-linier: Digunakan jika hubungan antara variabel tidak linier tetapi monotonic.
Kuat terhadap Outliers: Lebih robust terhadap outliers dibandingkan Pearson.
Data Kecil: Lebih baik digunakan pada dataset yang lebih kecil atau ketika data memiliki banyak nilai yang sama (ties).
3. Korelasi Spearman
* Tujuan: Mengukur hubungan monotonic antara dua variabel menggunakan peringkat.
* Rumus: Koefisien korelasi Spearman

* Nilai: Bervariasi antara -1 dan 1.
* Kondisi Penggunaan yang Cocok:
Data Berperingkat atau Skala Ordinal: Cocok untuk data ordinal atau data yang dapat diperingkat.

## Cara mengecek caracteristic dataset dengan visual grafik dan parameter lain

1. cek hubungan visual antara dua variabel yang berkorelatif dengan scatter plot


````python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.linear_model import LinearRegression
from sklearn.metrics import r2_score

# Contoh Data
df = pd.read_csv('california_housing_test.csv')


# Scatter Plot
sns.scatterplot(x='total_bedrooms', y='total_rooms', data=df)
plt.show()
````

* note:
1. Hubungan Linier Positif: Titik-titik data cenderung membentuk garis lurus yang naik dari kiri bawah ke kanan atas.
2. Hubungan Linier Negatif: Titik-titik data cenderung membentuk garis lurus yang turun dari kiri atas ke kanan bawah.
3. Tidak Linier: Titik-titik data tersebar tanpa membentuk pola garis lurus.

2. cek koefisien korelasi dan determinasi (R^2), kita fokuskan ke method pearson sbg contoh

````python
# Koefisien Korelasi, misal untuk method pearson
correlation = df.corr(method='pearson')
correlation

# Analisis Regresi Linier
X = df[['total_rooms']].values
Y = df['total_bedrooms'].values
model = LinearRegression()
model.fit(X, Y)
Y_pred = model.predict(X)
r2 = r2_score(Y, Y_pred)
print(f"Koefisien Determinasi (R^2): {r2}")
````

3. Residual plot untuk mengetahui pola

````python
# Residual Plot
residuals = Y - Y_pred
sns.residplot(x=Y_pred, y=residuals, lowess=True)
plt.xlabel('Predicted')
plt.ylabel('Residuals')
plt.show()
````

note:
Menganalisis plot residual (selisih antara nilai aktual dan nilai yang diprediksi oleh model regresi):

1. Tidak Ada Pola: Jika plot residual tidak menunjukkan pola tertentu dan tersebar secara acak di sekitar garis nol, ini mendukung adanya hubungan linier. selain itu, biasanya garisnya lurus.
2. Pola Tertentu: Jika ada pola dalam plot residual, ini menunjukkan adanya hubungan non-linier.

* Contoh kasus data ordinal

````python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Generate sample data
np.random.seed(0)
x = np.linspace(0, 10, 100)
y = np.sin(x) + np.random.normal(0, 0.1, 100)  # Non-linear monotonic relationship

df = pd.DataFrame({'x': x, 'y': y})

# Scatter plot
sns.scatterplot(x='x', y='y', data=df)
plt.show()

# Calculate correlations
pearson_corr = df.corr(method='pearson')
kendall_corr = df.corr(method='kendall')
spearman_corr = df.corr(method='spearman')

print(f"Pearson correlation:\n{pearson_corr}\n")
print(f"Kendall correlation:\n{kendall_corr}\n")
print(f"Spearman correlation:\n{spearman_corr}\n")
````
