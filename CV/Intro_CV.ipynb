### About inovation from AlexNet
1. 3 penemu dari AlexNet mempublish sebuah paper dalam lomba yang diadakan image net(pemilik dataset khusus cv yang salah satu terbesar dan terpopuler). paper tersebut berisi temuan terbaru mengenai deep learning concpet for cv dengan nama convolitional NN(CNN) architecture. Penemuan tersebut menunjukan concpet deep learning jadi outperform dari tradiotional(ML methods). Intinya, Alex net ini lebih mengandalkan GPUs dalam training model secara lebih cepat daripada CPUs. Penemuan ini pun menghasilkan penemuan lain yang inovativ dan terkenal hingga sekarang seperti penggunaan ReLu agar training lebih cepat, dropout regularization(mengurangi overfitting), overlapping max pooling yang membantu feature extraction, dan data augmentation untuk kebutuhan img processing.
