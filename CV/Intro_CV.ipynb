Feature Description: Ini harusnya juga ada dalam deep learning, sebenernya ga bisa dengan tepat disebut feature extractor(istilah dalam deep learning) karena outputnya adalah feature map(bukan descriptor yang siap dipakai atau hasil akhir untuk masuk ke tahap di luar feature extraction) yang merupakan hasil awal berupa map(struktur spasial + channel dari matrik). baru ketika nanti map selesai di tahap feature extraction jika dalam deep learning, disebutnya global descriptor(satu vektor untuk keseluruhan input layer NN).
2. Feature Descriptor: hasil representasi vektor yang mendeskripsikan suatu bagian gambar. Walaupun kesanya ini semacam alat/pelaku, dalam bahasa inggris ini tetep kata benda jadi merupakan suatu hasil. Dia beda dengan feature map karena bukan merupakan hasil awal dalam proses feature extraction.
3. Feature Matching:fokus proses membandingkan feature descriptor dari dua atau lebih gambar untuk menemukan kesamaan. Bayangkan ada dua gambar, kalau dilibrary openCV pakai tekni kayak rans itu pemasanganya direpresentasikan dengan garis-garis untuk menemukan kesamaan diantara dua gambar
4. Segmentation: pembagian gambar menjadi bagian-bagian bermakna
6. Global Descriptor: vektor ringkas yang mewakili keseluruhan input (hasil akhir setelah pooling/flatten) dalam proses feature extraction sebelum diolah ke fully connected layer(NN).
7. Local descriptor: representasi fitur pada patch tertentu (bagian mata, telinga, dan patch/bagian tertentu dalam gambar). Bedanya dengan global, dia ngambil map dalam deep learning itu di feature map sebelum pooling layer

### Filter
1. keknya dalam img yang di lakukan perhitungan dengan filter harus dipecah dulu bagian-bagian kecilnya(ex: mata/telinga/hidung) buat menentukan nilai hasil perhitungan tersebut nantinya ga sih?
* Ada betulnya. Itu disebut cara menghasilkan local sift(feature map) dari setiap output convolutional layer/convolutional filter.
2. Apakah artinya feature map dalam convolution layer ini merupakan edge? atau malah bagian/bahan-bahan pembentuk edge yang mana feature map terakhir sebelum masuk ke pooling layer/layer lainya?
* Feature map adalah hasil konvolusi antara input (misalnya gambar) dengan kernel/filter.
* Artinya, feature map ini adalah respons dari pola tertentu yang bisa berupa edge(gradien horizontal & vertical, tekstur, atau pola kompleks lain) yang diformat dalam bentuk matematis seperti matriks. Jadi, feature map diawal layer itu bukan langsung edge, tapi bagian dari penyusun edge alias response terhadap pola pixel gambar. Layar menengah membentuk edge semakin jelas dan kompleks, barulah setelah terlihat jelas terbentuk edge sebelum akhirnya masuk ke layer lain: pooling/dll.
2. Apa itu convolution filter? kenapa ada filter?
* Convolutional filter = convolutional layer.
* Filter yang dimaksud sama kayak kernel yang membantu feature extraction pada data image yang tersusun atas pixel-pixel. Lebih panjangnya dia itu implementasi dari perhitungan konvolusi mateimatis atau Approximate Finite Difference yang tujuanya untuk mencari edge atau perubahan intensitas pixel yakni magnitudenya dan arahnya.
3. Kenapa harus peduli tentang filter yang merupakan traditional method in cv,   padahal deep learning udah mengambil alih computer vision area?
* Jawaban: walaupun deep learning lebih sering dipakai, cv tradisional terkait filter/kernel ini punya history yang sampai sekarang di deep learning pasti kepakai dong alias cikal bakal. Deep learning dalam cv yang paling terkenal kan CNN, itupun dia tetep butuh yang namanya convolutional layer. Convolutional inilah yang ada implementasi filter/kernel. Jd tetep kepakai walaupun yang viral dan sering dipakai deep learning. Oleh karena itu, perlu banget paham sejarah/asal usul adanya si kernel ini di cv tradisional.
