# Creating an array

```python
# import library
import numpy as np
```

```python
# Perkalian dan dot product
a1 = np.array([[1, 2], [2, 4]])
a2 = np.array([[0, 1], [1, 0]])
print("Perkailan")
print(a1 * a2)
print("Dot product")
np.dot(a1, a2)
```
