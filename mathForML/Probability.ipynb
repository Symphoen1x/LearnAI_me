```python
# contoh referensi:



import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from scipy.stats import bernoulli

# Parameter untuk percobaan
p = 0.7  # Probabilitas sukses (sembuh)
sample_size = 1000  # Jumlah percobaan klinis

# Generate Bernoulli distribution
hasil_pengobatan = bernoulli.rvs(size=sample_size, p=p)

# Print the first few results to understand the data
print("Hasil percobaan klinis (10 percobaan pertama):", hasil_pengobatan[:10]) #melihat dan membaca proporsi jumlah pasien sembuh dan tidak sembuh

# Membuat plot visualiasi
sns.histplot(hasil_pengobatan, bins=2, kde=False)
plt.xlabel('Hasil Pengobatan')
plt.ylabel('Frekuensi')
plt.title('Distribusi Hasil Pengobatan dengan sample size {}'.format(sample_size))
plt.xticks([0, 1], ['Tidak Sembuh', 'Sembuh'])
plt.show()

# Melihat persebaran value dalam variabel hasil_pengobatan
unique, counts = np.unique(hasil_pengobatan, return_counts=True)
distribution = dict(zip(unique, counts))

print("Distribusi hasil pengobatan:", distribution)
print("Proporsi pasien yang sembuh:", distribution[1] / sample_size)
print("Proporsi pasien yang tidak sembuh:", distribution[0] / sample_size)
```

## Distribusi Binomial

* Implementasi dengan method rvs() dengan <br>
n=10: Jumlah percobaan (misalnya, 10 kali lemparan koin). <br>
p=0.5: Probabilitas sukses dalam setiap percobaan (misalnya, 
 probabilitas mendapatkan "kepala" pada setiap lemparan koin adalah 0.5).<br>
size=1000: Jumlah sampel yang dihasilkan (1000 kali eksperimen 10 lemparan koin).

```python
from scipy.stats import binom
import seaborn as sns
import matplotlib.pyplot as plt

koin = binom.rvs(size = 1000, n = 10, p = 0.5)

#Membuat plot
sns.histplot(koin)
plt.show()
```

* Implementasi dengan method pmf() untuk menhitung probabilitas pelemparan koin 10 kali. berapa peluang mendapatkan sisi gambar tepat 3 kali?

```python
import scipy.stats as stats

# Parameter distribusi binomial
n = 10  # jumlah percobaan (hari)
p = 0.5  # probabilitas sukses (hujan pada suatu hari)

# Menghitung probabilitas hujan turun tepat 5 kali (k = 5)
k = 3
probabilitas = stats.binom.pmf(k, n, p)

print(f"Probabilitas hujan turun tepat {k} kali dalam {n} hari adalah {probabilitas:.4f}")
```
